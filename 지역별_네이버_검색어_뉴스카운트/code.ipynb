```python
# 먼저 해당 라이브러리를 받아옵니다.
import requests
from bs4 import BeautifulSoup
import json
import re
import sys
import time, random
from random import randint
from time import sleep
import pandas as pd
```

```python
# 시군구 정보
city_list = pd.read_csv("./data/시군구.txt",header=None)
city_list = city_list[0].tolist()

# 시설정보
faci_list = pd.read_csv("./data/환경시설.txt",header=None)
faci_list = faci_list[0].tolist()

# 환경피해
damage_list = pd.read_csv("./data/환경피해.txt",header=None)
damage_list = damage_list[0].tolist()
```

```python
#header 추가
header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
}
```

```python
fd_list = []
for f in faci_list:
    for d in damage_list:
        temp = f + " " + d
        fd_list.append(temp)
```

```python
fd_list
```

```python
def get_news_cnt(q,s_date,e_date):  
    cnt = 0
    entry = 1
    s_from = s_date.replace(".","")
    e_to = e_date.replace(".","")
        
    while(True):
        url = "https://search.naver.com/search.naver?where=news&query=" + q + "&sort=0&ds=" + s_date + "&de=" + e_date + "&nso=so%3Ar%2Cp%3Afrom" + s_from + "to" + e_to + "%2Ca%3A&start=" + str(entry)  
        #print(url)
        breq = requests.get(url,headers=header)
        bsoup = BeautifulSoup(breq.content, 'html.parser')

        # html 파싱
        title_list = bsoup.select('div.news_wrap.api_ani_send > div > a')
        temp_cnt = len(title_list)
        #print(temp_cnt)
        cnt = cnt + temp_cnt
        if(temp_cnt < 10 or cnt >= 1000):
            return cnt
        entry = entry + 10
```

```python
result_df = pd.DataFrame(index = city_list, columns=fd_list)
s_date = "20170101"
e_date = "20220228"

for i in range(0,len(city_list)):
#for i in range(0,5):
    print(i)
    for j in range(0,len(fd_list)):
    #for j in range(0,5):
        q = city_list[i] + " " + fd_list[j] 
        cnt = get_news_cnt(q,s_date,e_date)
        result_df.iloc[i,j] = cnt
    sleep(5 + randint(1,100)/50)
    
```
